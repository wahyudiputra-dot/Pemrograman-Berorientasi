class adalah
sebuah cetak biru
atau template untuk membuat suatu objek.

```python
# Mendefinisikan suatu class untuk membangun suatu objek untuk mobil

class Mobil:
  jenis_mobil = 'Fortuner'
  warna = 'Silver'


print(Mobil)
```

Hasil bantuan dari GPT, yaitu menambahkan sebuah object : Mobilku = Mobil().
Agar dapat memunculkan objek dari class tersebut, seperti contoh pada program dibawah ini.

```python
class Mobil:
  jenis_mobil = "Fortuner"
  warna = "Silver"

Mobilku = Mobil()

print(Mobilku.jenis_mobil)
print(Mobilku.warna)
```

Sementara itu, untuk menginisialisasi sebuah objek, kita dapat menetapkan kelas yang telah didefinisikan.

```python
# Membuat objek yang berupa "Mobil_gua" dari class mobil
Mobil_gua = Mobil()

# Melakukan print untuk melihat spesifikasi "Mobil_gua"
print("---Informasi Mobil Gua dan Speknya---")
print("Jenis mobil gua :", Mobil_gua.jenis_mobil)
print("Warna mobil gua :", Mobil_gua.warna)
```
