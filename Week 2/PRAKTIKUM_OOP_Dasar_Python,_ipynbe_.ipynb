```python
kecepatan = 0
kecepatan_maksimal = 120
program_berjalan = True

while True:
    print("\n=== MENU KENDARAAN ===")
    print("1. Tambah Kecepatan")
    print("2. Kurangi Kecepatan")
    print("3. Lihat Status Kendaraan")
    print("4. Keluar Program")

    pilihan = input("Masukkan pilihan (1-4): ")

    if pilihan == "1":
        tambah = int(input("Masukkan jumlah kecepatan yang ingin ditambahkan: "))
        kecepatan += tambah

        if kecepatan > 120:
            kecepatan = 120
            print("Kecepatan mencapai batas maksimal 120 km/jam.")

        print("Kecepatan sekarang:", kecepatan, "km/jam")

    elif pilihan == "2":
        kurang = int(input("Masukkan jumlah kecepatan yang ingin dikurangi: "))
        kecepatan -= kurang

        if kecepatan < 0:
            kecepatan = 0

        print("Kecepatan sekarang:", kecepatan, "km/jam")

    elif pilihan == "3":
        print("Kecepatan kendaraan:", kecepatan, "km/jam")

        if kecepatan == 0:
            print("Kendaraan berhenti")
        else:
            print("Kendaraan berjalan")

    elif pilihan == "4":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid")
```
