# **Operators**

**Operator aritmatika**

Penjelasan : Program ini melakukan berbagai operasi aritmatika pada dua variabel yaitu penjumlahan, pengurangan, perkalian, pembagian, sisa bagi, pangkat, dan pembagian bulat, lalu menampilkan hasilnya menggunakan print().

```python
a = 10
b = 60

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
print(a ** b)
print(a // b)
```

**Operator Penugasan**

Penjelasan : Program ini menggunakan operator penugasan untuk mengubah nilai variabel, seperti += untuk menambahkan nilai, //= untuk melakukan pembagian bulat, dan &= untuk operasi logika AND pada nilai variabel, kemudian hasilnya ditampilkan dengan print().

```python
x = 30
x += 3
print(x)
```

```python
y = 70
y //= 2
print(y)
```

```python
x = 10
x &= 3
print(x)
```

**Operator Perbandingan**

Penjelasan : Program ini digunakan untuk membandingkan dua nilai menggunakan operator perbandingan seperti sama dengan, tidak sama dengan, lebih besar, lebih kecil, lebih besar sama dengan, dan lebih kecil sama dengan, lalu menampilkan hasilnya berupa nilai True atau False.

```python
a = 20
b = 10

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

**Operator Logika**

Penjelasan : Program ini menggunakan operator logika and, or, dan not untuk mengevaluasi kondisi dari dua variabel, kemudian menampilkan hasilnya berupa nilai True atau False.

```python
x = 7
y = 10

print(x > 5 and y < 15)
print(x > 10 or y < 15)
print(not(x > 5))
```

**Operator Keanggotaan**

Penjelasan : Program ini menggunakan operator keanggotaan in dan not in untuk memeriksa apakah suatu nilai terdapat di dalam list buah atau tidak, kemudian menampilkan hasilnya berupa nilai True atau False.

```python
buah = ["anggur", "sirsak", "semangka"]
print("anggur" in buah)
print("ayam" not in buah)
```

Penjelasan : Program ini menambahkan elemen baru ke dalam list fruits menggunakan method append() untuk menambah "Orange" di akhir list dan insert() untuk menambah "Grapes" pada indeks ke-1, kemudian menampilkan isi list tersebut menggunakan print().

Penjelasan : Program ini menghapus elemen dari list fruits dengan beberapa cara yaitu menggunakan remove() untuk menghapus elemen berdasarkan nilai, pop() untuk menghapus elemen berdasarkan indeks, dan del untuk menghapus elemen pada indeks tertentu, kemudian menampilkan isi list setelah perubahan menggunakan print().

# **Konversi Tipe Data**

Konversi Otomatis (Implicit Conversion)

Penjelasan: Program ini menjumlahkan angka_int dengan 5.5. Karena salah satu angkanya float, hasilnya otomatis menjadi float (55.5). Kemudian nilai dan tipenya ditampilkan dengan print().

```python
angka_int = 50
angka_float = angka_int + 5.5

print(angka_float)
print(type(angka_float))
```

Python mengubah 10 (int) menjadi 10.0 (float) agar sesuai dengan operasi float.

Konversi Manual (Explicit Conversion)

int ke float / float ke int

```python
x = 15
x_float = float(x)

y = 9.9
y_int = int(y)

print(x_float)
print(y_int)
```

int/float ke string

```python
angka = 53
angka_str = str(angka)

teks = "100"
teks_int = int(teks)

print(angka_str, type(angka_str))
print(teks_int, type(teks_int))
```

list ke tuple

```python
data_list = [1, 2, 3]
data_tuple = tuple(data_list)
data_tuple2 = (4, 5, 6)
data_list2 = list(data_tuple2)

print(data_tuple, type(data_tuple))
print(data_list2, type(data_list2))
```

list/tuple ke set

```python
angka_list = [1, 2, 2, 3]
angka_set = set(angka_list)

angka_tuple = tuple(angka_set)

print(angka_set)
print(angka_tuple)
```

Dictionary ke Lis

```python
data_dict = {"nama": "Wahyudi Menyala", "umur": 17}

keys_list = list(data_dict.keys())

values_list = list(data_dict.values())

print(keys_list)
print(values_list)
```

Kesimpulan:

1. Konversi Otomatis terjadi saat Python perlu menyamakan tipe data dalam operasi.
2. Konversi Manual (Casting) dilakukan dengan fungsi bawaan (int(), float(), str(), list(), dll.).
3. Gunakan konversi yang tepat sesuai kebutuhan, terutama untuk menghindari error saat operasi antar tipe data.
