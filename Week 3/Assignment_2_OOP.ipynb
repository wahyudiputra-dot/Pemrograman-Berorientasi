# **Buat suatu kelas bernama user**

```python
# Membuat kelas User
class User:
    def __init__(self, username, kelas, nim, email, umur):
        self.username = username
        self.kelas = kelas
        self.nim = nim
        self.email = email
        self.umur = umur

    def info_user(self):
        return f"Username: {self.username}, Kelas: {self.kelas}, Nim: {self.nim}, Email: {self.email}, Umur: {self.umur}"

# Membuat objek dari kelas User
Wahyudi = User("Wahyudi", "Robotika Pagi C", 4222501086, "wahyudiarigata6@gmail.com", "19 Tahun")
Iqbal = User("Iqbal", "Robotika Pagi A", 42225010806, "iqbalganteng@gmail.com", "19 Tahun")

# Menampilkan informasi pengguna
print(f"""
--- Informasi Wahyudi ---
Username: {Wahyudi.username}
Kelas: {Wahyudi.kelas}
Nim: {Wahyudi.nim}
Email: {Wahyudi.email}
Umur: {Wahyudi.umur}
"""
)

print(f"""
--- Informasi Iqbal ---
Username: {Iqbal.username}
kelas: {Iqbal.kelas}
Nim: {Iqbal.nim}
Email: {Iqbal.email}
Umur: {Iqbal.umur}
"""
)
```
